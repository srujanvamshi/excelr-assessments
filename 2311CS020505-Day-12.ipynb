```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the specified path
file_path = r"C:/Users/bhanu/Downloads/Day_12_banking_data.csv"  # Use 'r' to avoid backslash issues
df = pd.read_csv(file_path)

# Bar plot using Pandas: Total sum of Transaction_Amount per Account_Type
transaction_sum_per_account = df.groupby('Account_Type')['Transaction_Amount'].sum()
transaction_sum_per_account.plot(kind='bar', 
                                 title='Total Transaction Amount per Account Type',
                                 xlabel='Account Type', 
                                 ylabel='Total Transaction Amount',
                                 color='skyblue', 
                                 rot=45)

# Show the bar plot
plt.tight_layout()
plt.show()

# Pie chart using Pandas: Percentage of transactions per Branch
transaction_count_per_branch = df['Branch'].value_counts()
transaction_count_per_branch.plot(kind='pie', 
                                  autopct='%1.1f%%', 
                                  title='Percentage of Transactions per Branch')

# Show the pie chart
plt.tight_layout()
plt.show()
```
